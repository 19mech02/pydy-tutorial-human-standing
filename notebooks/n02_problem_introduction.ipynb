In this tutorial we are going to make a simple model of a human (or humaniod robot) that is capable of balancing on its own. We will make several assumptions:

- The motion is limited to a 2D plane.
- We only have three degrees of freedom: ankle, knee, and hip flexion/extension.
- The left ankle, knee and hip joints change angle the same as the right joints.
- The muscles will be modeled as ideal torques between the body segments.

Reference Frames
----------------

There are four reference frames and six important points. The orange inertial reference frame, $I$, is attached to the foot which is rigidly attached to the ground. The blue lower leg reference frame, $L$, is attached to the foot by a pin joint at the angkle point $A$ and rotates relative to the foot through $\theta_1$. The green upper leg reference frame, $U$, is attached to the lower leg by a pin joint at the knee point $K$ and rotates relative to the lower leg through angle $\theta_2$. The red torsoe reference frame, $T$, is pinned to the upper leg at the hip point, $H$, and rotates relative to the upper leg through the angle $\theta_3$.

Mass Centers
-------------

There are three points at the mass center of each body segment: $L_o$, $U_o$, $T_o$. These are each located on the line connecting the proximal and distal joints of each body segment and are located by the dimensions: $d_L$, $d_U$, and $d_T$.

Gravity
-------

Gravity is directed downwards and applies a force with a magnitude of $mg$ at each mass center.

Torques
-------

Three torques represent the forces due to muscles contracting and extending. The ankle $T_A$, knee $T_K$, and hip $T_H$ torques apply equal and opposite torques to the adjoining body segments.
